# Behavioral Shift Detection & Pattern Transition Analysis

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

import warnings
warnings.filterwarnings("ignore")
```

```

df = pd.read_csv("sales_06_FY2020-21.csv")
df.head()
```

```

df.info()
```

```

df_clean = df.dropna().drop_duplicates()
df_clean.isnull().sum()
```

```

df_clean['Date'] = pd.to_datetime(df_clean['Date'])
df_clean['Month'] = df_clean['Date'].dt.month
df_clean['Year'] = df_clean['Date'].dt.year
```

```

features = ['Sales', 'Quantity', 'Profit']
X = df_clean[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```

sns.histplot(df_clean['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()
```

```

monthly_sales = df_clean.groupby('Month')['Sales'].sum()
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.show()
```

```

kmeans = KMeans(n_clusters=3, random_state=42)
df_clean['Behavior_Cluster'] = kmeans.fit_predict(X_scaled)
```

```

sns.scatterplot(x='Sales', y='Profit', hue='Behavior_Cluster', data=df_clean)
plt.title("Behavioral Clusters")
plt.show()
```

```

pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)

df_clean['PCA1'] = pca_components[:,0]
df_clean['PCA2'] = pca_components[:,1]
```

```

sns.scatterplot(x='PCA1', y='PCA2', hue='Behavior_Cluster', data=df_clean)
plt.title("Pattern Transition (PCA)")
plt.show()
```

```

X_ml = df_clean[['Sales', 'Quantity', 'Profit']]
y_ml = df_clean['Behavior_Cluster']

X_train, X_test, y_train, y_test = train_test_split(
    X_ml, y_ml, test_size=0.2, random_state=42
)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
